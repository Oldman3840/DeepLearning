```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow
from tensorflow import keras
from tensorflow.keras.utils import to_categorical

import numpy as np

from sklearn.model_selection import train_test_split
```

```python
###############################
# train set / test set
# 조건 1. test set size = 10%
# 조건 2. 난수 고정 = 2023
###############################
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.1, random_state= 2023)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
iris.target_names
```

```python
class_n = len(np.unique(y_train))
class_n
```

```python
#######################
# 어떤 전처리 작업 필요: One-hot encoding
# Hint : y.shape
#######################
y_train = to_categorical(y_train, class_n)
y_test = to_categorical(y_test, class_n)
```

```python
y_train.shape, y_test.shape
```

```python
########
# 모델링
########
# 1. 세션 종료
keras.backend.clear_session()

# 2. 모델 선언
model = keras.models.Sequential()

# 3. 모델 블록 조립
model.add(keras.layers.Input(4, ))
model.add(keras.layers.Dense(3, activation = 'softmax'))

# 4. 컴파일
model.compile(loss = 'categorical_crossentropy', metrics= ['accuracy'], optimizer= 'adam')
model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs = 10, verbose= 1)
```

```python
import pandas as pd
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
```

# 히든 레이어 추가하여 모델링! (Functional API)
---
- 조건
    - 히든 레이어 2개를 추가하여 다시 모델링!
    - 히든 레이어 노드 수 : 32
    - 히든 레이어 activation = 'relu'

```python
# 라이브러리 호출
import tensorflow
from tensorflow import keras
from tensorflow.keras.utils import to_categorical

from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
# 데이터 나누기
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 2023)
```

```python
iris.feature_names, iris.target_names
```

```python
class_n = len(np.unique(y_train))
```

```python
# 데이터 전처리
y_train = to_categorical(y_train, class_n)
y_test = to_categorical(y_test, class_n)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
########
# 모델링: ANN
########
# 1. 세션 클리어
keras.backend.clear_session()

# 2. 모델 선언
model = keras.models.Sequential()

# 3. 모델 블록 조립
model.add(keras.layers.Input(shape = (4, )))
model.add(keras.layers.Dense(32, activation = 'relu'))
model.add(keras.layers.Dense(32, activation = 'relu'))
model.add(keras.layers.Dense(3, activation = 'softmax'))
          
# 4. 컴파일
# model.compile(loss = 'categorical_crossentropy', metrics= ['accuracy'], optimizer= 'adam')
model.compile(loss = keras.losses.categorical_crossentropy , metrics= ['accuracy'], optimizer= 'adam')
model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs = 10, verbose = 1)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
print(y_test[:10])
print(y_pred[:10])
```
