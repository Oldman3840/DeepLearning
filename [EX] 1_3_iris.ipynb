```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
y
```

```python
iris.target_names
```

```python
# One-Hot Encoding
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y, 3)
```

```python
x.shape, y.shape
```

## 1번. Sequential API style로 모델링 하세요

- loss = 'categorical_crossentropy' 사용!

```python
import tensorflow as tf
from tensorflow import keras

import pandas as pd
import numpy as np
```

```python
# 모델링: Sequential
# 1. 세선 클리어
keras.backend.clear_session()

# 2. 모델 선언
model = keras.models.Sequential()

# 3. 모델 블록 설정
model.add(keras.layers.Input(shape= (4, )))
model.add(keras.layers.Dense(3, activation = 'softmax'))

# 4. 컴파일
# loss = 'categorical_crossentropy' : multi class일 때 반드시 사용
# model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy', optimizer = 'adam')
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy', optimizer = keras.optimizers.Adam(0.1))
```

```python
model.fit(x, y, epochs = 20, verbose= 1)
```

```python
model.summary()
```

```python
cols_name = ['setosa', 'versicolor', 'virginica']
result = pd.DataFrame.from_records(y, columns=cols_name)
```

```python
result[['p_setosa', 'p_versicolor', 'p_virginica']] = model.predict(x)
result
```
